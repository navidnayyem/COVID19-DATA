**12 Outliers are above 1233 and correspond date is shown on the y axis**

```python
limit = 1233 # maximum limit of boxplot
date_column = dataset2['Reporting Date'] # assign Reporting Date column from dataset into date_column variable
inf_column = dataset2['Infected (Last 24 Hours)'] # assign Infected (Last 24 Hours) column from dataset into inf_column variable
count_outlier_values = inf_column[inf_column > limit].count() # count the total number of outlier values of Infected (Last 24 Hours)
inf_outlier_date = date_column[inf_column > limit] # find infected outlier reporting date whose limit is greater than 11500 
inf_outlier_values = inf_column[inf_column > limit] # find infected outlier values whose limit is greater than 11500 

print('Total Outlier Values of Infection from 12Sept21 - 9Jan22:', count_outlier_values) # print the total number of outlier values of Infected (Last 24 Hours)
print('\nReporting Date      | Value') # print heading
print('-----------------------------') # print broken line
table_output = "\n".join("{} | {}".format(x, y) for x, y in zip(inf_outlier_date, inf_outlier_values)) # joining table
print(table_output) # print table output of outlier values of Infected (Last 24 Hours) after joining 
```

**First Quartile(Q1), Median, Third Quartile(Q3), Interquartile Range(IQR),Maximum & Minimum of Infection**

```python
minimum = min(dataset1['Infected (Last 24 Hours)']) # formula to find minimum value of Deaths
q1 = dataset1['Infected (Last 24 Hours)'].quantile(0.25) #percentile25(1st quartile of Deaths)
q3 = dataset1['Infected (Last 24 Hours)'].quantile(0.75) #percentile75(3rd quartile of Deaths)
iqr = q3 - q1 #formula of interquartile range of Deaths
lower_limit = (q1 - (1.5 * iqr)) # formula of lower limit of Deaths
upper_limit = (q3 + (1.5 * iqr)) # formula of upper limit of Deaths
maximum = max(dataset1['Infected (Last 24 Hours)']) # formula to find maximum value of Deaths

print('Lower Limit: ',lower_limit) # print lower limit of Deaths
print('Minimum Value: ',minimum) # print minimum value of Deaths
print('First Quartile(Q1): ',q1) #print 1st quartile of Deaths
print('Third Quartile(Q3): ',q3) #print 3rd quartile of Deaths
print('Upper Limit: ',upper_limit) # print upper limit of Deaths
print('Maximum: ',maximum) # print maximum value of Deaths
print('Interquartile Range(IQR): ',iqr) #print interquartile range of Deaths 
```
